```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python

ax=sns.countplot(x='Gender',data=df,hue='Gender')
# ax.set_xticks(range(len(ax.get_xticklabels())))  
# ax.set_xticklabels(["Female Customer","Male Customer"])
ax.set_ylabel('Number of customer ')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sale_gen=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sale_gen
```

```python
sns.barplot(x="Gender",y="Amount",data=sale_gen,hue="Gender")
```

By analysing above both plot we can say the number of female buyser is greater than male buyers and then purchasing power of female is higher than male as we can see total amount spent by male and female

After analyzing the data, we gained insight that unmarried females aged 26–35 working in the IT, Aviation, and Healthcare sectors from Uttar Pradesh, Maharashtra, and Karnataka spent more money on food, clothing, and electronic gadgets during Diwali.
